```python
import pandas as pd

data = pd.read_csv("carData.csv", index_col=0)
data = data[data["TOTAL"] > 499]
data.sort_values("TOTAL", inplace=True, ascending=False)
total_cars = data.sum()["TOTAL"]
total_makes = data.sum()[1:]
data
```

```python
temp_data = []
for make in total_makes.index:
    temp_data.append(total_makes[make] / total_cars)
makes_base = pd.DataFrame(temp_data, columns=["ratio"], index=total_makes.index)
makes_base
```

```python
ratio_data = pd.DataFrame(index=data.index, columns=data.columns[1:])

for suburb in data.index:
    for make in ratio_data.columns:
        ratio_data.loc[suburb, make] = (data.loc[suburb, make] / data.loc[suburb, "TOTAL"]) / makes_base.loc[make, "ratio"]
ratio_data.to_csv("ratioData.csv", index_label="suburb")
ratio_data
```

```python
car_suburbs = pd.DataFrame(index=ratio_data.index, columns=["top_make", "ratio"])
for index, row in ratio_data.iterrows():
    car_suburbs.loc[index, "top_make"] = row.sort_values(ascending=False).index[0]
    car_suburbs.loc[index, "ratio"] = row.sort_values(ascending=False)[0]
car_suburbs.sort_index(inplace=True)
car_suburbs.to_csv("car_suburb.csv", index_label="suburb")
car_suburbs
```
