```python
import pandas as pd

data = pd.read_csv("carData.csv", index_col=0)
data = data[data["TOTAL"] > 0]
data
```

```python
# sexuality data
data_sex = pd.read_csv("./data/suburbSexuality.csv", header=6, usecols=[0,1,2,3])
data_sex.columns = ["suburb", "male_only", "female_only", "opposite"]
data_sex["suburb"] = data_sex["suburb"].apply(lambda x: x.upper())
data_sex.set_index("suburb", inplace=True)
data_sex.index.name = None
data_sex.dropna(inplace=True)
data_sex["total"] = data_sex.sum(axis=1)
data_sex["male_only"] = data_sex["male_only"].astype("float")
data_sex
```

```python
def isNaN(value):
    return value != value

makes = data.columns
sex_data = pd.DataFrame(index=makes, columns=["male_only", "female_only", "opposite"])
for make in makes:
    values = [
        ["male_only", 0],
        ["female_only", 0],
        ["opposite", 0]
    ]
    print("Preparing", make, "data ...")
    for suburb in data[make].index:
        temp_suburb = suburb
        if not suburb in data_sex.index:
            temp_suburb = suburb + " (ACT)"
        for index, value in enumerate(values):
            temp_value = data_sex.loc[temp_suburb, value[0]] / data_sex.loc[temp_suburb, "total"]
            if not isNaN(temp_value):
                values[index][1] = values[index][1] + data[make].loc[suburb] * temp_value
    for index, value in enumerate(values):
        sex_data.loc[make, value[0]] = value[1]
    print(make, "analysis complete")
sex_data
```

```python
final_sex = pd.DataFrame(index=sex_data.index)
for sexuality in sex_data.columns:
    final_sex[sexuality] = sex_data[sexuality] / sex_data.sum(axis=1)
final_sex.sort_values("male_only", ascending=False)
```
