```python
import pandas as pd

data = pd.read_csv("vehicles.csv", usecols=["make", "suburb", "count"])

unique_suburbs = data["suburb"].unique()
unique_makes = data["make"].unique()

suburbs = data.groupby("suburb").sum()
suburbs.sort_values("count", inplace=True, ascending=False)
suburbs = suburbs[suburbs["count"] > 149]
suburbs.index.name = None

new_data = pd.DataFrame(index=suburbs.index)

suburbs
```

```python
makes = data.groupby("make").sum()
makes.sort_values("count", inplace=True, ascending=False)
makes = makes[makes["count"] > 499]
makes.index.name = None
makes
```

```python
for make in makes.index:
    print("Preparing", make, "data ...")
    make_count = []
    for suburb in suburbs.index:
        make_count.append(data[(data["suburb"] == suburb) & (data["make"] == make)]["count"].sum())
    new_data[make] = make_count
    print(make, "data complete")
```

```python
new_data["TOTAL"] = new_data.sum(axis=1)
new_data.to_csv("carData.csv")
```
