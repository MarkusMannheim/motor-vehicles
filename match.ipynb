```python
import pandas as pd

car_data = pd.read_csv("carData.csv")
car_data["suburb"] = car_data["suburb"].apply(lambda x: x.upper())
car_data.set_index("suburb", inplace=True)
car_data.index.name = None
car_data
```

```python
suburb_data = pd.read_csv("suburbs.csv", usecols=["SSC_NAME_2016", "STATE_NAME_2016"])
suburb_data.columns = ["suburb", "state"]
suburb_data["suburb"] = suburb_data["suburb"].apply(lambda x: x.upper().replace(" (ACT)", ""))
suburb_data.set_index("suburb", inplace=True)
suburb_data.index.name = None
suburb_data
```

```python
for suburb in car_data.index:
    result = suburb in suburb_data.index
    if not result:
        print(suburb, result)
```
